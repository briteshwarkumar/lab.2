```python
import numpy as np

# Sample Data
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# Initialize weights and biases
w1 = np.random.randn(2, 3)
b1 = np.zeros((1, 3))
w2 = np.random.randn(3, 1)
b2 = np.zeros((1, 1))

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Assuming x, y, w1, b1, w2, b2 are already initialized
epochs = 1000
learning_rate = 0.1
loss_history = []

for i in range(epochs):
    # --- Step 5: Forward Propagation ---
    z1 = np.dot(x, w1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, w2) + b2
    a2 = sigmoid(z2)

    # --- Step 6: Compute Loss ---
    loss = np.mean((y - a2)**2)
    loss_history.append(loss)

    # --- Step 7: Backpropagation ---
    error_output = a2 - y
    d_weights2 = np.dot(a1.T, error_output * (a2 * (1 - a2)))

    error_hidden = np.dot(error_output * (a2 * (1 - a2)), w2.T)
    d_weights1 = np.dot(x.T, error_hidden * (a1 * (1 - a1)))

    # --- Step 8: Update Weights and Biases ---
    w2 -= learning_rate * d_weights2
    b2 -= learning_rate * np.sum(error_output * (a2 * (1 - a2)), axis=0, keepdims=True)
    w1 -= learning_rate * d_weights1
    b1 -= learning_rate * np.sum(error_hidden * (a1 * (1 - a1)), axis=0, keepdims=True)

    if i % 100 == 0:
        print(f"Epoch {i}, Loss: {loss}")
```

```python
import matplotlib.pyplot as plt

# 1. Plot loss vs epochs
plt.figure(figsize=(10, 5))
plt.plot(loss_history)
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.grid(True)
plt.show()

# 2. Plot activation function (Sigmoid)
def plot_sigmoid():
    sample_z = np.linspace(-10, 10, 100)
    plt.plot(sample_z, sigmoid(sample_z))
    plt.title('Sigmoid Activation Function')
    plt.show()

plot_sigmoid()
```

```python
# 1. Provide a test input
# For example, if training an XOR gate, let's test [1, 0]
test_input = np.array([[1, 0]])

# 2. Perform forward propagation (using the trained weights w1, w2, b1, b2)
# Hidden Layer
z1_test = np.dot(test_input, w1) + b1
a1_test = sigmoid(z1_test)

# Output Layer
z2_test = np.dot(a1_test, w2) + b2
a2_test = sigmoid(z2_test)

# 3. Display predicted output
print(f"Test Input: {test_input}")
print(f"Predicted Output: {a2_test}")

# Optional: Round to nearest integer for classification
print(f"Final Classification: {np.round(a2_test)}")
```
